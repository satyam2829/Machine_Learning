### Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading the data

```python
data = pd.read_csv('iris.csv')
```

### Top 5 rows

```python
data.head()
```

### Checking null values

```python
data.isnull().sum()
```

<h3 style='color:red'> Observation: There are no missing values in the data.</h3>

### Species Count

```python
data['Species'].value_counts()
```

<h3 style='color:red'> Observation: The dataset is balanced. All the species are distributed equally.</h3>

### Countplot: Species

```python
sns.countplot(data['Species'])
plt.title('Iris Species Count')
plt.show()
```

# 1. Univariate Analysis

## 1.1. Histogram

```python
sepal_length = data['SepalLengthCm']
sepal_width = data['SepalWidthCm']
petal_length = data['PetalLengthCm']
petal_width = data['PetalWidthCm']
```

### 1.1.1. Sepal Length

```python
plt.hist(sepal_length,edgecolor='black', bins=10)
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.title('Histogram - Sepal Length')
plt.show()
```

### 1.1.2. Sepal Width

```python
plt.hist(sepal_width,edgecolor='black', bins=10)
plt.xlabel('Sepal Width')
plt.ylabel('Count')
plt.title('Histogram - Sepal Width')
plt.show()
```

### 1.1.3. Petal Length

```python
plt.hist(petal_length,edgecolor='black', bins=10)
plt.xlabel('Petal Length')
plt.ylabel('Count')
plt.title('Histogram - Petal Length')
plt.show()
```

### 1.1.4. Petal Width

```python
plt.hist(petal_width,edgecolor='black', bins=10)
plt.xlabel('Petal Length')
plt.ylabel('Count')
plt.title('Histogram - Petal Width')
plt.show()
```

## 1.2. Boxplot

### 1.2.1. Sepal Length

```python
sns.boxplot(x='Species',y='SepalLengthCm',data=data)
plt.show()
```

### 1.2.2. Sepal Width

```python
sns.boxplot(x='Species',y='SepalWidthCm',data=data)
plt.show()
```

### 1.2.3. Petal Length

```python
sns.boxplot(x='Species',y='PetalLengthCm',data=data)
plt.show()
```

### 1.2.4. Petal Width

```python
sns.boxplot(x='Species',y='PetalWidthCm',data=data)
plt.show()
```

# 2.Bivariate Analysis

## 2.1. Scatter Plot

### 2.1.1. Sepal Length vs Sepal Width

```python
sns.scatterplot(sepal_length, sepal_width, hue=data['Species'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

### 2.1.2. Petal Length vs Petal Width

```python
sns.scatterplot(petal_length, petal_width, hue=data['Species'])
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Sepal Length vs Sepal Width')
plt.legend(loc='lower right')
plt.show()
```

# 3. Correlation

```python
cor_mat = df_corr.corr()
```
